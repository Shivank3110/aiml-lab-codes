```python
from collections import deque
graph = {
    'a' : ['b', 'c'],
    'b' : ['d', 'e', 'f'],
    'c' : ['e', 'f'],
    'd' : ['e', 'f'],
    'e' : ['f'],
    'f' : []
}

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    print(" bfs traversal ")
    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node, end="\n")
            visited.add(node)
            for neighbour in graph[node]:
                if neighbour not in visited:
                    queue.append(neighbour)
bfs(graph, 'a')
```

```python
from collections import deque
graph = {
    'a' : ['b', 'c'],
    'b' : ['d', 'e', 'f'],
    'c' : ['e', 'f'],
    'd' : ['e', 'f'],
    'e' : ['f'],
    'f' : []
}
visited = set()
def dfs(graph,node):
    if node not in visited:
        print(node,end = "\n" )
        visited.add(node)
        for neighbour in graph[node]:
            dfs(graph, neighbour)
print(" dfs traversal ")
dfs(graph, 'a')
```
